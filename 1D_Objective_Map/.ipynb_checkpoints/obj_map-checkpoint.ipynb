```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def obj(x) :
    # return 0.025*np.square(x) + np.sin(x)
    # return np.abs(x)
    return np.exp(np.cos(np.square(x))) + np.square(x)

def objder(x) :
    # return 0.05*x + np.cos(x)
    # return 1 if x > 0 else -1
    return 2*x*(1 + np.sin(np.square(x)) * np.exp(np.cos(np.square(x))))
```

```python
llim = -20
rlim = 20

xch = []
x = np.random.rand()*40 - 20

xch.append(x)
for _ in range(0, 100) :
    x = x - objder(x)/10
    if not (llim < x < rlim) :
        print ("Out of bounds!")
        break
    xch.append(x)
```

```python
print ("Start at {} --> f = {}, df/dx = {}".format(xch[0], obj(xch[0]), objder(xch[0])))
print ("End at {} --> f = {}, df/dx = {}".format(xch[-1], obj(xch[-1]), objder(xch[-1])))
```

```python
fig, ax = plt.subplots(1,2,gridspec_kw={'width_ratios' : [1,1]})
fig.set_figheight(5)
fig.set_figwidth(10)

ax[0].plot (xch, list(map(obj, xch)))
ax[1].plot (xch, list(map(objder, xch)))
```
